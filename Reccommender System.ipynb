```python
df['Start Year']=df['Start Year'].str.replace('II','').str.replace('I','').str.strip()
```

```python
df['Start Year'].value_counts()
```

```python
df['Start Year']=df['Start Year'].astype('int')
```

```python
def start_year(x):
    if x<2000:
        return 1
    elif x>=2000 and x<2020:
        return 2
    else:
        return 3
df['Start Year']=df['Start Year'].apply(start_year)
```

```python
main_genre=[]
for i in df['Genre'].str.split(','):
    main_genre.append(i[0])
```

```python
df['Main_Genre']=main_genre
```

```python
other_genre=[]
for i in df['Genre'].str.split(','):
    if len(i)>1:
        other_genre.append(i[1])
    else:
        other_genre.append(np.nan)
```

```python
df['Other_Genre']=other_genre
df.info()
```

```python
df.drop(columns='Genre',inplace=True)
```

```python
def replacement(x):
    if x=='Musical':
        return 'Music'
    else:
        return x
df['Main_Genre']=df['Main_Genre'].apply(replacement)
df['Main_Genre'].value_counts()
```

```python
df['Other_Genre']=df['Other_Genre'].str.strip()
df['Other_Genre']=df['Other_Genre'].apply(replacement)
df['Other_Genre'].value_counts()
```

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(strategy='most_frequent')
df[['Other_Genre']]=imputer.fit_transform(df[['Other_Genre']])
df
```

```python
df['Overview']=df['Overview'].str.lower()
import string
def rem_punc(text):
    exclude = set(string.punctuation) # returns a set of all punctuations in text
    return ''.join(ch for ch in text if ch not in exclude)
df['Overview'] = df['Overview'].apply(rem_punc)
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
sw=set(stopwords.words('english'))
def remove_stopwords(text):
    text_tokens = text.split()
    filtered_tokens = [word for word in text_tokens if word not in sw]
    return ' '.join(filtered_tokens)
df['Overview']=df['Overview'].apply(remove_stopwords)
```

```python
 import nltk
from nltk.stem import PorterStemmer
nltk.download('punkt')
stemmer = PorterStemmer()
def stem_text(text):
    tokens = nltk.word_tokenize(text)
    stemmed_tokens = [stemmer.stem(token) for token in tokens]
    return ' '.join(stemmed_tokens)
df['Overview'] = df['Overview'].apply(stem_text)
```

```python
m=0
for i in df['Overview']:
    m=np.maximum(m,len(i))
m
```

```python
import gensim.downloader as api
word2vec_model = api.load('word2vec-google-news-300')
```

```python
def document_vector(doc):
    doc = [word for word in doc.split() if word in word2vec_model]
    return np.mean(word2vec_model[doc], axis=0)
overview_vectors = df['Overview'].apply(document_vector)
```

```python
overview_vectors = np.array(overview_vectors.tolist()) 
overview_vectors
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=50)
overview_vectors=pca.fit_transform(overview_vectors)
overview_vectors
```

```python
ov=pd.DataFrame(overview_vectors,index=df.index)
df=pd.concat([df,ov],axis=1)
df.drop(columns='Overview',inplace=True)
df
```
