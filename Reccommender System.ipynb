```python
df['Start Year']=df['Start Year'].str.replace('II','').str.replace('I','').str.strip()
```

```python
df['Start Year'].value_counts()
```

```python
df['Start Year']=df['Start Year'].astype('int')
```

```python
def start_year(x):
    if x<2000:
        return 1
    elif x>=2000 and x<2020:
        return 2
    else:
        return 3
df['Start Year']=df['Start Year'].apply(start_year)
```

```python
main_genre=[]
for i in df['Genre'].str.split(','):
    main_genre.append(i[0])
```

```python
df['Main_Genre']=main_genre
```

```python
other_genre=[]
for i in df['Genre'].str.split(','):
    if len(i)>1:
        other_genre.append(i[1])
    else:
        other_genre.append(np.nan)
```

```python
df['Other_Genre']=other_genre
df.info()
```

```python
df.drop(columns='Genre',inplace=True)
```

```python
def replacement(x):
    if x=='Musical':
        return 'Music'
    else:
        return x
df['Main_Genre']=df['Main_Genre'].apply(replacement)
df['Main_Genre'].value_counts()
```

```python
df['Other_Genre']=df['Other_Genre'].str.strip()
df['Other_Genre']=df['Other_Genre'].apply(replacement)
df['Other_Genre'].value_counts()
```

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(strategy='most_frequent')
df[['Other_Genre']]=imputer.fit_transform(df[['Other_Genre']])
df
```
