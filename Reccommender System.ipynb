```python
df['Overview']=df['Overview'].str.lower()
import string
def rem_punc(text):
    exclude = set(string.punctuation) # returns a set of all punctuations in text
    return ''.join(ch for ch in text if ch not in exclude)
df['Overview'] = df['Overview'].apply(rem_punc)
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
sw=set(stopwords.words('english'))
def remove_stopwords(text):
    text_tokens = text.split()
    filtered_tokens = [word for word in text_tokens if word not in sw]
    return ' '.join(filtered_tokens)
df['Overview']=df['Overview'].apply(remove_stopwords)
```

```python
 import nltk
from nltk.stem import PorterStemmer
nltk.download('punkt')
stemmer = PorterStemmer()
def stem_text(text):
    tokens = nltk.word_tokenize(text)
    stemmed_tokens = [stemmer.stem(token) for token in tokens]
    return ' '.join(stemmed_tokens)
df['Overview'] = df['Overview'].apply(stem_text)
```

```python
import gensim.downloader as api
word2vec_model = api.load('word2vec-google-news-300')
```

```python
def document_vector(doc):
    doc = [word for word in doc.split() if word in word2vec_model]
    return np.mean(word2vec_model[doc], axis=0)
overview_vectors = df['Overview'].apply(document_vector)
```

```python
overview_vectors = np.array(overview_vectors.tolist()) 
overview_vectors
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=100)
overview_vectors=pca.fit_transform(overview_vectors)
overview_vectors
```

```python
ov=pd.DataFrame(overview_vectors,index=df.index)
df=pd.concat([df,ov],axis=1)
df.drop(columns='Overview',inplace=True)
df
```

```python
df['Genre']=df['Genre'].str.replace(',','')
```

```python
df['Star1']=df['Star1'].str.strip().str.lower()
df['Star2']=df['Star2'].str.strip().str.lower()
df['Star3']=df['Star3'].str.strip().str.lower()
df['Star4']=df['Star4'].str.strip().str.lower()
df['Genre']=df['Genre'].str.strip().str.lower()
df['Series_Title']=df['Series_Title'].str.strip().str.lower()
```

```python
def concatenation(x):
    return ''.join(x.split(' '))
df['Star1']=df['Star1'].apply(concatenation)
df['Star2']=df['Star2'].apply(concatenation)
df['Star3']=df['Star3'].apply(concatenation)
df['Star4']=df['Star4'].apply(concatenation)
```

```python
df['stars']=df['Star1']+" "+df['Star2']+" "+df['Star3']+" "+df['Star4']
df.drop(columns=['Star1','Star2','Star3','Star4'],inplace=True)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vectorizer = CountVectorizer(stop_words='english', max_features=100)
count_matrix = count_vectorizer.fit_transform(df['Genre'])
count_df = pd.DataFrame(count_matrix.toarray(), columns=count_vectorizer.get_feature_names_out(), index=df.index)
df = pd.concat([df, count_df], axis=1)
df.drop(columns='Genre',inplace=True)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer(stop_words='english',max_features=100)
tfidf_matrix = tfidf_vectorizer.fit_transform(df['stars'])
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out(),index=df.index)
df=pd.concat([df,tfidf_df],axis=1)
df
```
